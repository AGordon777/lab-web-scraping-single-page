Let's cluster based on the numerical acoustic data.

I'll use StandardScaler to normalise the dataset.

```python
X_num = song_library[['danceability', 'energy', 'key',
       'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness',
       'liveness', 'valence', 'tempo', 'duration_ms', 'time_signature']]
X_num.dtypes
```

```python
from sklearn.preprocessing import StandardScaler

X_prep = StandardScaler().fit_transform(X_num)
pd.DataFrame(X_prep,columns=X_num.columns).describe().T
```

```python
X_num.head()
```

### Next: check for optimal number of clusters

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import cluster
from sklearn.cluster import KMeans
```

Let's test for the optimal number of clusters.

Ideally we want the smallest possible n_clusters that gives the lowest inertia or highest silhouette score.

```python
from tqdm import tqdm
```

```python
%%time

K = range(2, 20)
inertia = []

for k in tqdm(K):
    kmeans = KMeans(n_clusters=k,
                    random_state=1234, n_init=10)
    kmeans.fit(X_prep)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

Maybe n_clusters = 10? Hard to spot the elbow.

We'll try a silhouette score:

```python
%%time # this one takes a while!

from sklearn.metrics import silhouette_score
K = range(2, 20)
silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234, n_init=10)
    kmeans.fit(X_prep)
    silhouette.append(silhouette_score(X_prep, kmeans.predict(X_prep)))


plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Slhouette score showing the optimal k')
```

So there are 9 clusters in the song_library dataset based on acoustic features.

### Clustering the dataset

I'll use kmeans as a method for clustering the data and for predicting the cluster of a future user-picked song.

```python
kmeans = KMeans(n_clusters=9, random_state=1234) # k=9, found earlier
kmeans.fit(X_prep)

clusters = kmeans.predict(X_prep)

pd.Series(clusters).value_counts().sort_index()
```

```python
display(len(song_library))
len(clusters)
```

```python
song_library['category'] = clusters
song_library.head()
```
